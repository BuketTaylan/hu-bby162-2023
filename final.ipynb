```python
#Buket TAYLAN Katalog Projesi 

from google.colab import drive
drive.mount('/gdrive')

katalog = []

dosya_yolu = "/gdrive/My Drive/Colab Notebooks/veri.txt"
 
def menu_goster():
    print("""
    Katalog Uygulaması
    1. Eserleri Listele
    2. Yeni Eser Ekle
    3. Eser Sil
    4. Eser Güncelle
    5. Katalogu Temizle
    6. Eser Ara
    7. Çıkış
    """)


def eserleri_listele():
  with open(dosya_yolu, "r", encoding="utf-8") as dosya:

    for satir in dosya:
      veri = satir.strip()
      eser_bilgileri = veri.split(" - ")

      if len(eser_bilgileri) == 3:
        eser_adi = eser_bilgileri[0]
        yazar_adi = eser_bilgileri[1]
        basim_yili = eser_bilgileri[2]
        print("Eser Adı:", eser_adi)
        print("Yazar Adı:", yazar_adi)
        print("Basım Yılı:", basim_yili)
        print("-----------------------")
      else:
        print("Hatalı veri formatı.")
  dosya.close()
def yeni_eser_ekle():
    eser = {
        "Eser Adı": input("Eser Adı: "),
        "Yazar Adı": input("Yazar Adı: "),
        "Basım Yılı": input("Basım Yılı: ")
    }
    katalog.append(eser)
    with open(dosya_yolu, "a", encoding="utf-8") as dosya:
      veri = f"{eser['Eser Adı']} - {eser['Yazar Adı']} - {eser['Basım Yılı']}\n"
      dosya.write(veri)
    
    print("Yeni eser kataloga eklendi.")
    dosya.close()

def eser_sil():
    eser_adi = input("Silinecek Eser Adı: ")

    with open(dosya_yolu, "r", encoding="utf-8") as dosya:
        katalog = []
        for satir in dosya:
            veri = satir.strip()
            if veri:
                eser_bilgileri = veri.split(" - ")
                if len(eser_bilgileri) == 3:
                    eser = {
                        "Eser Adı": eser_bilgileri[0],
                        "Yazar Adı": eser_bilgileri[1],
                        "Basım Yılı": eser_bilgileri[2]
                    }
                    katalog.append(eser)
    dosya.close()

    silinecek_eserler = [eser for eser in katalog if eser["Eser Adı"] == eser_adi]
    if silinecek_eserler:
        katalog = [eser for eser in katalog if eser["Eser Adı"] != eser_adi]
        with open(dosya_yolu, "w", encoding="utf-8") as dosya:
            for eser in katalog:
                veri = f"{eser['Eser Adı']} - {eser['Yazar Adı']} - {eser['Basım Yılı']}\n"
                dosya.write(veri)
        print("Eser katalogdan ve dosyadan silindi.")
        dosya.close()
    else:
        print("Belirtilen eser katalogda bulunamadı.")


def eser_guncelle():
    eser_adi = input("Güncellenecek Eser Adı: ")
    with open(dosya_yolu, "r", encoding="utf-8") as dosya:
        katalog = []
        for satir in dosya:
            veri = satir.strip()
            if veri:
                eser_bilgileri = veri.split(" - ")
                if len(eser_bilgileri) == 3:
                    eser = {
                        "Eser Adı": eser_bilgileri[0],
                        "Yazar Adı": eser_bilgileri[1],
                        "Basım Yılı": eser_bilgileri[2]
                    }
                    katalog.append(eser)
    dosya.close()

    
    guncellenecek_eserler = [eser for eser in katalog if eser["Eser Adı"] == eser_adi]
    
    if guncellenecek_eserler:
        for eser in guncellenecek_eserler:
            print("Eski Bilgiler:")
            print("Eser Adı:", eser["Eser Adı"])
            print("Yazar Adı:", eser["Yazar Adı"])
            print("Basım Yılı:", eser["Basım Yılı"])

            eser["Eser Adı"] = input("Yeni Eser Adı: ")
            eser["Yazar Adı"] = input("Yeni Yazar Adı: ")
            eser["Basım Yılı"] = input("Yeni Basım Yılı: ")

        with open(dosya_yolu, "w", encoding="utf-8") as dosya:
            for eser in katalog:
                veri = f"{eser['Eser Adı']} - {eser['Yazar Adı']} - {eser['Basım Yılı']}\n"
                dosya.write(veri)
        print("Eser bilgileri güncellendi.")
        dosya.close()
    else:
        print("Belirtilen eser katalogda bulunamadı.")


def katalogu_temizle():
    katalog.clear()
    
    onay = input("Katalogu tamamen temizlemek istediğinizden emin misiniz? (E/H): ")
    
    if onay.lower() == "e":
      with open(dosya_yolu, 'w') as dosya:
        dosya.write("")
      print("Katalog tamamen temizlendi.") #Buket TaYLAN
      dosya.close()
    else:
      print("İşlem iptal edildi.")

def eser_ara():
    aranan = input("Aranacak Kelime: ")
    with open(dosya_yolu, "r", encoding="utf-8") as dosya:
      katalog = []
      for satir in dosya:
        veri = satir.strip()
        if veri:
          eser_bilgileri = veri.split(" - ")
          if len(eser_bilgileri) == 3:
            eser = {
                "Eser Adı": eser_bilgileri[0],
                "Yazar Adı": eser_bilgileri[1],
                "Basım Yılı": eser_bilgileri[2]
            }
            katalog.append(eser)

    bulunan_eserler = []
    for eser in katalog:
        if aranan.lower() in eser["Eser Adı"].lower() or aranan.lower() in eser["Yazar Adı"].lower():
            bulunan_eserler.append(eser)

    if bulunan_eserler:
        print("Eser Bulundu:")
        for eser in bulunan_eserler:
            print("Eser Adı:", eser["Eser Adı"])
            print("Yazar Adı:", eser["Yazar Adı"])
            print("Basım Yılı:", eser["Basım Yılı"])
            print("----------")
    else:
        print("Belirtilen kelimeye uygun eser bulunamadı.")

    dosya.close()


while True:
    menu_goster()
    secim = input("Bir seçenek girin (1-7): ")
    #BukeT TAYLAN
    if secim == "1":
        eserleri_listele()
    elif secim == "2":
        yeni_eser_ekle()
    elif secim == "3":
        eser_sil()
    elif secim == "4":
        eser_guncelle()
    elif secim == "5":
        katalogu_temizle()
    elif secim == "6":
        eser_ara()
    elif secim == "7":
        print("Katalog uygulaması sonlandırılıyor.")
        break
    else:
        print("Geçersiz bir seçenek girdiniz. Lütfen tekrar deneyin.")
```
